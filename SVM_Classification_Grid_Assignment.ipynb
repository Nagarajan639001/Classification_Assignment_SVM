```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#adding the Dada Set
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
#Splite input & Out put data set
```

```python
dataset.columns
```

```python
independant=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
independant
```

```python
dependant=dataset[["classification_yes"]]
```

```python
dependant
```

```python
# split the Train set & test set dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =train_test_split(independant,dependant,test_size=1/3,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

```

```python
scy= StandardScaler()
Y_train=scy.fit_transform(Y_train)
Y_test=scy.transform(Y_test)
```

```python
X_test
```

```python
# Train data-Model creation
```

```python
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV
param_grid= {'kernel':['rbf','ploy','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auti','scale']}
grid= GridSearchCV(SVC(probability=True),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')

# Fitting the Model for grid seach 
grid.fit(X_train,Y_train)
```

```python
# Evaluation Method Presidcting the Test set result
# print best parameter after tuning  
#print(grid.best_params_) 
re=grid.cv_results_
#pint(re)
```

```python
grid_predictions=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,grid_predictions)
```

```python
# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test,grid_predictions)
```

```python
print("The Confusion_matrix :\n",cm )
```

```python
print("The Report: \n",clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
